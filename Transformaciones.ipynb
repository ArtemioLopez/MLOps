```python
import pandas as pd
import numpy as np
import ast
```

```python
#Cargamos los datasets
df_movies = pd.read_csv('datasets\movies_dataset.csv')
df_credits = pd.read_csv('datasets\credits.csv')
```

```python
#Visualizamos los datos de movies
df_movies
```

## Merge


```python
df_movies.head(2)
```

```python
#Identificamos la cantidad de ids duplicados en movies y en credits
print(df_movies.duplicated(subset='id').sum(), df_credits.duplicated(subset='id').sum())
```

```python
#Quitamos los duplicados 
df_movies.drop_duplicates(subset='id', keep='first', inplace=True)
df_credits.drop_duplicates(subset='id', keep='first', inplace=True)
```

```python
print(df_movies.shape, df_credits.shape)
```

```python
#Eliminamos estos registros que tienen datos faltantes y deesplazados
df_movies.drop([35587,19730,29503], axis=0, inplace=True)
```

```python
df_movies.shape
```

```python
#cambiamos el formato de la columna id 
df_movies['id'] = df_movies['id'].astype(int)
```

```python
df_movies['id'].dtype
```

```python
#Realizamos el merge
df_final = pd.merge(df_movies, df_credits, on='id', how='inner')
```

```python
df_final.head(3)
```

```python
df_final.shape
```

## Tratamos las columna anidadas

### Belongs to collection

```python
df_final['belongs_to_collection']
```

```python
#Se evalua el valor de la celda y se transforma en un diccionario, para luego acceder a la clave name y poder extraerlo, SI el valor es nulo se asigna None
df_final['belongs_to_collection'] = df_final['belongs_to_collection'].apply(lambda x: ast.literal_eval(x)['name'] if pd.notnull(x) else None)
```

```python
df_final['belongs_to_collection']
```

### Genres

```python
df_final['genres']
```

```python
df_final['genres'] = df_final['genres'].apply(ast.literal_eval)
```

```python
df_final['genres'] = df_final['genres'].apply(lambda x: ', '.join([d['name'] for d in x]) if x else None)
```

```python
df_final['genres']
```

### Production companies

```python
df_final['production_companies']
```

```python
df_final['production_companies'] = df_final['production_companies'].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else None)
```

```python
df_final['production_companies'] = df_final['production_companies'].apply(lambda x: ', '.join([d['name'] for d in x]) if x else None)
```

```python
df_final['production_companies']
```

### Production Countries

```python
df_final['production_countries']
```

```python
#Verificamos si el registro es una cadena de texto, si es asi se convierte en un objeto de python
df_final['production_countries'] = df_final['production_countries'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else None)
```

```python
df_final['production_countries'] = df_final['production_countries'].apply(lambda x: ', '.join([d['name'] for d in x]) if isinstance(x, list) else "Sin Dato")
```

```python
df_final['production_countries']
```

### Spoken Languages

```python
df_final['spoken_languages']
```

```python
df_final['spoken_languages'] = df_final['spoken_languages'].apply(lambda x: ast.literal_eval(x) if isinstance(x,str) else x)
```

```python
df_final['spoken_languages'] = df_final['spoken_languages'].apply(lambda x: ', '.join([d['name'] for d in x]) if isinstance(x,list) else None)
```

```python
df_final['spoken_languages']
```

## Imputamos valores Nulos de las columnas revenue, budget.

```python
df_final.isnull().sum()
```

```python
#Transformamos en 0 los valores nulos de la columna revenue
df_final['revenue'].fillna(0, inplace=True)
```

```python
df_final.isnull().sum()
```

## Tranformamos el formato para la columna release_date. Va a tener el siguiente patron AAAA-mm-dd

```python
df_final['release_date']
```

```python
#Eliminamos los valores nulos de la columna release_date
df_final.dropna(subset=['release_date'], inplace=True)
```

```python
#Transformamos la columna en un objeto de fecha y hora
df_final['release_date'] = pd.to_datetime(df_final['release_date'])
```

```python
df_final['release_date']
```

```python
#Nos quedamos de release_date, solo con el año para conformar la nueva columna 
df_final['release_year'] = df_final['release_date'].dt.year
```

```python
df_final['release_year']
```

## Creamos la columna "return". 

```python
df_final['revenue']
```

```python
#Cambiamos el formato de la columna 
df_final['revenue'] = df_final['revenue'].astype(int)
```

```python
df_final['revenue'].dtype
```

```python
#Cambiamos el formato de la columna budget
df_final['budget'] = df_final['budget'].astype(int)
```

```python
df_final['budget'].dtype
```

```python
#Si budget es distinto de 0, dividimos el valor de la columna budget por el valor de la columna revenue. Para conformar la columna "return"
df_final['return'] = np.where(df_final['budget'] != 0, df_final['revenue'] / df_final['budget'], 0)
```

```python
df_final['return']
```

## Eliminamos las columnas que no seran utilizadas

```python
df_final = df_final.drop(['video','imdb_id','adult', 'original_title','poster_path','homepage'], axis=1)
```

```python
df_final.head(2)
```

## Desanidamos las columnas "crew" y "cast"

### Cast

```python
df_final['cast']
```

```python
df_final['cast'] = df_final['cast'].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else None)
```

```python
df_final['cast'][0]
```

```python
#Desanidamos la columna cast tomando los primeros 5 nombres de cada reparto 
df_final['cast'] = df_final['cast'].apply(lambda x: ', '.join([d['name'] for d in x[:5]]) if x else None)
```

```python
df_final['cast']
```

### Crew

```python
df_final['crew'][0]
```

```python
df_final['crew'] = df_final['crew'].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else None)
```

```python
df_final['crew'][0]
```

```python
df_final['Director'] = df_final['crew'].apply(lambda x: ', '.join([d['name'] for d in x if d.get('job') == 'Director']) if isinstance(x,list) else "")
```

```python
df_final['Director']
```

```python
# Eliminamos la columna crew
df_final.drop('crew', axis=1, inplace=True)
```

```python
df_final.head()
```

## Cambiamos el formato de la columna Runtime

```python
df_final['runtime'].fillna(0, inplace=True)
```

```python
df_final['runtime'] = df_final['runtime'].astype(int)
```

```python
df_final['runtime']
```

```python
#Para Eda
df_final.to_csv('datasets\Movies_Final.csv', index=False)
```
